```python
import pandas as pd
import numpy as np
import operator
import sys
```

```python
df_full = pd.read_csv("Data/data.csv")
```

```python
id = "3Fj47GNK2kUF0uaEDgXLaD"
df_id = df_full[(df_full.id == id)]
cluster_val = df_id['cluster'].iloc[0]
df_cluster = df_full[(df_full.cluster == cluster_val)]
features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']
df_id_features = df_id[features]
df_id_features_list = list(df_id_features.iloc[0])
df_cluster['distance'] = sys.float_info.max
df_cluster = df_cluster[(df_cluster.popularity > 40)]
df_cluster = df_cluster[df_cluster.year > 2015]
df_cluster = df_cluster.reset_index()
```

```python
n = len(df_cluster)
n
```

```python
df_features = df_cluster[features]
for i in range(n):
    df_cluster['distance'][i] = np.linalg.norm(np.array(list(map(operator.sub, list(df_features.iloc[i]), df_id_features_list))))
```

```python
df_cluster = df_cluster.sort_values('distance')
df_cluster
```

```python
df_cluster.to_csv("closest_distance1.csv", index=False)
```
